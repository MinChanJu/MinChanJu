```python
import turtle

def line(obj,a,b,th,r,k=0):
    obj.penup()
    obj.goto(a,b)
    obj.pendown()
    obj.setheading(th)
    if k == 0: obj.forward(r)
    else:
        while r > 0:
            obj.forward(14)
            obj.penup()
            obj.forward(6)
            obj.pendown()
            r-=20
    return 

def square(obj,a,b,th,w,h):
    obj.penup()
    obj.goto(a,b)
    obj.pendown()
    obj.setheading(th)
    obj.forward(w)
    obj.left(90)
    obj.forward(h)
    obj.left(90)
    obj.forward(w)
    obj.left(90)
    obj.forward(h)
    obj.left(90)
    return

def make_cicle(obj,a,b,th,r,t=0):
    obj.penup()
    obj.goto(a,b)
    obj.pendown()
    obj.setheading(th)
    if t==0: obj.circle(r)
    else: obj.circle(r,t)
    return

def window(obj,a,b,w,h):
    square(obj,a,b,0,w,h)
    line(obj,a,b+h//2,0,w)
    line(obj,a+w//2,b,90,h)
    return

def move(obj,a,b,th=0):
    obj.penup()
    obj.goto(a,b)
    obj.pendown()
    obj.setheading(th)
    return

# 터틀 화면 생성
screen = turtle.Screen()
screen.screensize(2000, 1000)

# 터틀 객체 생성
road = turtle.Turtle()
building = turtle.Turtle()
car = turtle.Turtle()
tree = turtle.Turtle()

# 터틀 객체 속도 설정
road.speed(20)
building.speed(20)
car.speed(20)
tree.speed(20)

# 밑 바닥
line(road,-400,-200,0,800)
line(road,-400,-215,0,800,5)
line(road,-400,-230,0,800)

# 건물
square(building,-300,-200,0,130,250)
line(building,-320,50,0,170)
building.goto(-170,70)
building.goto(-300,70)
building.goto(-320,50)
line(building,-300,65,0,130)
line(building,-305,60,0,140)
line(building,-310,55,0,150)
square(building,-200,70,0,10,15)
square(building,-278,-200,0,26,52)
make_cicle(building,-273,-175,0,2)

window(building,-280,0,30,30)
window(building,-220,0,30,30)
window(building,-280,-60,30,30)
window(building,-220,-60,30,30)
window(building,-280,-120,30,30)
window(building,-220,-120,30,30)
window(building,-220,-180,30,30)

# 자동차
line(car,-100,-190,0,30)
make_cicle(car,-50,-190,90,10,180)
make_cicle(car,-100,-190,90,10,180)
line(car,-50,-190,0,10)
line(car,-130,-190,0,10)
make_cicle(car,-40,-190,90,30,40)
make_cicle(car,-130,-190,-90,30,-40)
line(car,-123,-172,0,13)
line(car,-59,-172,0,13)
make_cicle(car,-110,-172,-90,50,-25)
make_cicle(car,-59,-172,90,50,25)
line(car,-105,-151,0,42)
make_cicle(car,-110,-200,0,8)
make_cicle(car,-59,-200,0,8)
line(car,-118,-192,0,16)
line(car,-110,-200,90,16)
line(car,-67,-192,0,16)
line(car,-59,-200,90,16)
move(car,-115,-187)
car.goto(-105,-197)
move(car,-64,-187)
car.goto(-54,-197)
move(car,-115,-197)
car.goto(-105,-187)
move(car,-64,-197)
car.goto(-54,-187)
square(car,-100,-172,0,31,15)

# 나무
move(tree,-3,-200)
tree.goto(0,-50)
tree.goto(3,-200)
for i in range(10):
    if i>5:
        move(tree,0,-50-10*i)
        tree.goto(-30-2*(10-i),-70-10*i)
        move(tree,0,-50-10*i)
        tree.goto(30+2*(10-i),-70-10*i)
        continue
    move(tree,0,-50-10*i)
    tree.goto(-30-2*i,-70-10*i)
    move(tree,0,-50-10*i)
    tree.goto(30+2*i,-70-10*i)


# 터틀 그래픽 창 열린 상태로 유지
turtle.done()
```

```python
from math import sqrt # sqrt(n) = n**(1/2)
def sieve_w(n):
    # 1단계
    lst = [i for i in range(2, n+1)]

    # 2단계
    p = 2
    # 3단계
    while p <= sqrt(n):
        lst = [x for x in lst if x == p or x % p != 0]
        # 4단계
        for x in lst:
            if x > p:
                p = x
                break
        else:
            break

    return lst
```

```python
def is_solvable(puzzle):
    """
    직사각형 슬라이딩 퍼즐이 풀 수 있는지 확인하는 함수
    :param puzzle: 2D 리스트로 표현된 퍼즐 배열
    :return: True (풀 수 있음), False (풀 수 없음)
    """
    rows = len(puzzle)  # 퍼즐의 행 개수
    cols = len(puzzle[0])  # 퍼즐의 열 개수

    # 퍼즐을 1D 리스트로 변환하고 빈 칸 위치 찾기
    flattened = []
    blank_row = 0  # 빈 칸이 있는 행 (맨 아래에서부터 계산)

    for i in range(rows):
        for j in range(cols):
            if puzzle[i][j] == 0:  # 0은 빈 칸
                blank_row = rows - i  # 행 번호를 맨 아래에서부터 계산
            else:
                flattened.append(puzzle[i][j])
    
    # 무질서도(inversions) 계산
    inversions = 0
    for i in range(len(flattened)):
        for j in range(i + 1, len(flattened)):
            if flattened[i] > flattened[j]:
                inversions += 1

    # 퍼즐의 가로 크기(cols)가 홀수인지 짝수인지에 따라 조건 확인
    if cols % 2 == 1:
        # 가로 크기가 홀수일 경우, 무질서도가 짝수여야 함
        return inversions % 2 == 0
    else:
        # 가로 크기가 짝수일 경우, (빈 칸의 행 번호 + 무질서도)가 홀수여야 함
        return (blank_row + inversions) % 2 == 1
```
