```python
def f(x) :
    return x*x*x + 4*(x*x) - 10 
```

```python
a = 1
b = 2
n = 1
p = (a+b)/2

while 1 :
    if 1/(2**n) < 10**(-3) :
        print(n, p, f(p))
        break
    
    if f(a)*f(p) < 0 :
        a = a
        b = p
        p = (a+b)/2
        n += 1
    elif f(b)*f(p) < 0 :
        a = p
        b = b
        p = (a+b)/2
        n += 1
```

```python
import numpy as np
```

```python
# Bisection Method
def f(x):
    return x-np.cos(x)

a0 = 0
b0 = 1
n = 0
p = (a0 + b0)/2

a = a0
b = b0

while (1):
    print(n, p)
    if f(a)*f(p)<0:
        a = a
        b = p
        p = (a + b)/2
        n = n + 1
        
    else:
        a = p
        b = b
        p = (a + b)/2
        n = n + 1
        
    if (b0-a0)/2**n < 10**(-10):
        break
        
print(n, p)
```

```python
# fixed point
def g(x):
    return np.cos(x)

n = 0
p0 = 0

while (1):
    p1 = g(p0)
    print(p1)
    n += 1
    
    if abs(p1-p0) < 10**(-10):
        break
    
    p0 = p1
```

```python
# newton's method
def g(x):
    return x - (x-np.cos(x))/(1+np.sin(x))

n = 0
p0 = 0

while (1):
    p1 = g(p0)
    print(n, p1)
    n += 1
    
    if abs(p1-p0) < 10**(-10):
        break
    
    p0 = p1
```

```python
# secant method
def g(x, pre_x):
    return x - ( (x-np.cos(x))*(x-pre_x) ) / ( (x-np.cos(x))-(pre_x-np.cos(pre_x)) )

n = 0
p0 = 0
pre_x = 1

while (1):
    p1 = g(p0, pre_x)
    print(n, p1)
    n += 1                  
    
    if abs(p1-p0) < 10**(-10):
        break
        
    pre_x = p0
    p0 = p1 
```

```python
# Flase position method
def g(x, pre_x):
    return x - ( (x-np.cos(x))*(x-pre_x) ) / ( (x-np.cos(x))-(pre_x-np.cos(pre_x)) )

n = 0
p0 = 0
pre_x = 1

while (1):
    p1 = g(p0, pre_x)
    print(n, p1)
    n += 1                  
    
    if abs(p1-p0) < 10**(-10):
        break
        
    p0 = p1 
```
