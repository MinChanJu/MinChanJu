# 먹이 생성

```python
import pygame

# Pygame 초기화
pygame.init()

# 게임 화면 설정
screen_width = 1000
screen_height = 1000
size = [screen_width,screen_height]
screen = pygame.display.set_mode(size)

# 제목 설정
title = 'PAC MAN'
pygame.display.set_caption(title)

# 게임 내 설정
clock = pygame.time.Clock()
color = (120,120,255)

# 객체 생성
class obj:
    # 초기 위치 설정
    def __init__(self):
        self.x = 0
        self.y = 0
    # 이미지 추가
    def put_img(self, address):
        self.img = pygame.image.load(address).convert_alpha()
        self.sx, self.sy = self.img.get_size()
    # 이미지 사이즈 변경
    def change_size(self, sx, sy):
        self.img = pygame.transform.scale(self.img,(sx,sy))
        self.sx, self.sy = self.img.get_size()
    # 객체 그리기 
    def show(self):
        screen.blit(self.img, (self.x, self.y))
 
# 팩맨 객체 생성       
pacman = obj()
pacman.put_img("./이미지/팩맨1 위쪽.png")
pacman.change_size(40,40)
pacman.x = (size[0]//2) - (pacman.sx//2)
pacman.y = (size[1]*2)//3
pacman.move = 8

# 팩맨 객체 이동방향 초기 값
left_go = False
right_go = False
up_go = False
down_go = False

# 팩맨 먹이 리스트 생성
point_list = []
for i in range(20, size[0], 60):
    for j in range(200, size[1], 60):
        point = obj()
        point.put_img("./이미지/점.png")
        point.change_size(10,10)
        point.x = i
        point.y = j
        point_list.append(point)

# 필요한 변수 선언
cnt = 0 # FPS 시간 계산
direct = 1 # 팩맨 입 열었는지 닫았는지 여부
key = [pygame.K_LEFT,pygame.K_RIGHT,pygame.K_UP,pygame.K_DOWN]
image = ["./이미지/팩맨1 왼쪽.png","./이미지/팩맨1 오른쪽.png","./이미지/팩맨1 위쪽.png","./이미지/팩맨1 아래쪽.png"]
image2 = ["./이미지/팩맨2 왼쪽.png","./이미지/팩맨2 오른쪽.png","./이미지/팩맨2 위쪽.png","./이미지/팩맨2 아래쪽.png"]

# 게임 루프
running = True
while running:

    # FPS 설정
    clock.tick(60)

    # event가 일어났는지 확인
    for event in pygame.event.get():
        # 창 X표시 누르면 게임 종료
        if event.type == pygame.QUIT:
            running = False
        # 키 입력 확인
        if event.type == pygame.KEYDOWN:
            for i in range(4):
                if event.key == key[i]:
                    go = [1 if i == j else 0 for j in range(4)]
                    left_go = bool(go[0])
                    right_go = bool(go[1])
                    up_go = bool(go[2])
                    down_go = bool(go[3])
                    pacman.put_img(image[i]) # 이미지 재설정
                    pacman.change_size(40,40)
                    direct = 1 # 픽맨 입 연것으로 재설정
    
    # FPS 시간 증가
    cnt += 1
    
    # FPS 시간 8마다 입 다물거나 열기
    for i in range(4):
        if [left_go,right_go,up_go,down_go][i] == True and cnt%8 == 0:
            if direct == 1:
                pacman.put_img(image2[i])
                pacman.change_size(40,40)
                direct = 2
            elif direct == 2:
                pacman.put_img(image[i])
                pacman.change_size(40,40)
                direct = 1
    
    # 각 이동 방향으로 이동
    if left_go == True:
        pacman.x -= pacman.move
        if pacman.x <= 0:
            pacman.x = 0
    if right_go == True:
        pacman.x += pacman.move
        if pacman.x >= size[0] - pacman.sx:
            pacman.x = size[0] - pacman.sx
    if up_go == True:
        pacman.y -= pacman.move
        if pacman.y <= 0:
            pacman.y = 0
    if down_go == True:
        pacman.y += pacman.move
        if pacman.y >= size[1] - pacman.sy:
            pacman.y = size[1] - pacman.sy

    # 화면 채우기
    screen.fill(color)
    
    # 팩맨 그리기
    pacman.show()
    
    # 남은 먹이들 그리기
    for p in point_list:
        p.show()
        
    # FPS 시간이 너무 커지는 것을 방지하기 위해 초기화
    if cnt == 88: cnt = 0

    # 화면을 매 프레임마다 업데이트
    pygame.display.flip()

pygame.quit()
```

```python
import pygame

# Pygame 초기화
pygame.init()

# 게임 화면 설정
screen_width = 1000
screen_height = 1000
size = [screen_width,screen_height]
screen = pygame.display.set_mode(size)

# 제목 설정
title = 'PAC MAN'
pygame.display.set_caption(title)

# 게임 내 설정
clock = pygame.time.Clock()
color = (120,120,255)

# 객체 생성
class obj:
    # 초기 위치 설정
    def __init__(self):
        self.x = 0
        self.y = 0
    # 이미지 추가
    def put_img(self, address):
        self.img = pygame.image.load(address).convert_alpha()
        self.sx, self.sy = self.img.get_size()
    # 이미지 사이즈 변경
    def change_size(self, sx, sy):
        self.img = pygame.transform.scale(self.img,(sx,sy))
        self.sx, self.sy = self.img.get_size()
    # 객체 그리기
    def show(self):
        screen.blit(self.img, (self.x, self.y))
        
# 충돌 감지
def crash(a, b):
    if (a.x-b.sx <= b.x <= a.x+a.sx):
        if (a.y-b.sy <= b.y <= a.y+a.sy):
            return True
    return False

# 게임 성공했을 때 사용할 배경 객체
game_success = obj()
game_success.put_img("./이미지/게임 성공.png")
game_success.change_size(size[0],size[1])
 
# 팩맨 객체 생성       
pacman = obj()
pacman.put_img("./이미지/팩맨1 위쪽.png")
pacman.change_size(40,40)
pacman.x = (size[0]//2) - (pacman.sx//2)
pacman.y = (size[1]*2)//3
pacman.move = 8

# 팩맨 객체 이동방향 초기 값
left_go = False
right_go = False
up_go = False
down_go = False

# 팩맨 먹이 리스트 생성
point_list = []
for i in range(20, size[0], 60):
    for j in range(200, size[1], 60):
        point = obj()
        point.put_img("./이미지/점.png")
        point.change_size(10,10)
        point.x = i
        point.y = j
        point_list.append(point)

# 필요한 변수 선언
cnt = 0 # FPS 시간 계산
direct = 1 # 팩맨 입 열었는지 닫았는지 여부
'''
time = 0 # 게임 시작했는지 여부
'''
playing = 1 # 게임이 끝났는지 여부
key = [pygame.K_LEFT,pygame.K_RIGHT,pygame.K_UP,pygame.K_DOWN]
image = ["./이미지/팩맨1 왼쪽.png","./이미지/팩맨1 오른쪽.png","./이미지/팩맨1 위쪽.png","./이미지/팩맨1 아래쪽.png"]
image2 = ["./이미지/팩맨2 왼쪽.png","./이미지/팩맨2 오른쪽.png","./이미지/팩맨2 위쪽.png","./이미지/팩맨2 아래쪽.png"]

# 게임 루프
running = True
while running:

    # FPS 설정
    clock.tick(60)

    # event가 일어났는지 확인
    for event in pygame.event.get():
        # 창 X표시 누르면 게임 종료
        if event.type == pygame.QUIT:
            running = False
        # 키 입력 확인
        if event.type == pygame.KEYDOWN and playing == 1:
            for i in range(4):
                if event.key == key[i]:
                    go = [1 if i == j else 0 for j in range(4)]
                    left_go = bool(go[0])
                    right_go = bool(go[1])
                    up_go = bool(go[2])
                    down_go = bool(go[3])
                    pacman.put_img(image[i]) # 이미지 재설정
                    pacman.change_size(40,40)
                    direct = 1 # 픽맨 입 연것으로 재설정
                    '''
                    time = 1 # 게임이 시작 된 것으로 설정
                    '''
    
    # FPS 시간 증가
    cnt += 1
    
    '''
    # 만약 게임이 시작되지 않았다면 초기 시간 초기화
    if time == 0:
        start_ticks = pygame.time.get_ticks()
        font = pygame.font.Font(None, 50)
        cnt = 0
    '''
    
    # FPS 시간 8마다 입 다물거나 열기
    for i in range(4):
        if [left_go,right_go,up_go,down_go][i] == True and cnt%8 == 0:
            if direct == 1:
                pacman.put_img(image2[i])
                pacman.change_size(40,40)
                direct = 2
            elif direct == 2:
                pacman.put_img(image[i])
                pacman.change_size(40,40)
                direct = 1
    
    # 각 이동 방향으로 이동
    if left_go == True:
        pacman.x -= pacman.move
        if pacman.x <= 0:
            pacman.x = 0
    if right_go == True:
        pacman.x += pacman.move
        if pacman.x >= size[0] - pacman.sx:
            pacman.x = size[0] - pacman.sx
    if up_go == True:
        pacman.y -= pacman.move
        if pacman.y <= 0:
            pacman.y = 0
    if down_go == True:
        pacman.y += pacman.move
        if pacman.y >= size[1] - pacman.sy:
            pacman.y = size[1] - pacman.sy

    # 화면 채우기
    screen.fill(color)
    
    # 팩맨 그리기
    pacman.show()
    
    # 지울 먹이 리스트를 만들어 팩맨과 먹이가 충돌했을 때(먹었을 때) 리스트에 추가
    remove_point_list = []
    for p in point_list:
        if crash(pacman,p):
            remove_point_list.append(p)

    # 지울 먹이 리스트에 있는 모든 먹이들을 제거
    for dp in remove_point_list:
        point_list.remove(dp)
    
    # 남은 먹이들 그리기
    for p in point_list:
        p.show()
    
    # FPS 시간이 너무 커지는 것을 방지하기 위해 초기화
    if cnt == 88: cnt = 0
    
    # 만약 먹이 리스트의 길이가 0일 경우 (모든 먹이를 먹었을 경우) 게임 성공 객체 그리고 게임 끝났는지 여부를 저장하는 playing 변수를 0으로 설정
    if len(point_list) == 0:
        game_success.show()
        playing = 0
        left_go = False
        right_go = False
        up_go = False
        down_go = False
        '''
        # 게임 진행 시간 표시
        screen.blit(time_text, (300, 750))
    
    # 만약 게임 중이라면 초를 다시 계산하여 표시
    if playing == 1:
        seconds = (pygame.time.get_ticks() - start_ticks) / 1000
        minute = int(seconds)//60
        seconds -= minute*60
        if minute == 0:
            time_text = font.render(f"Time : {seconds:.2f} sec", True, (255,255,255))
        else:
            time_text = font.render(f"Time : {minute:} min {seconds:.2f} sec", True, (255,255,255))
        screen.blit(time_text, (10, 10))
    '''

    # 화면을 매 프레임마다 업데이트
    pygame.display.flip()

pygame.quit()
```
