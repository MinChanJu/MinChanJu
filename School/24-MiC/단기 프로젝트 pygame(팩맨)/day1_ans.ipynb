# 팩맨 객체 생성

```python
import pygame

# Pygame 초기화
pygame.init()

# 게임 화면 설정
screen_width = 1000
screen_height = 1000
size = [screen_width,screen_height]
screen = pygame.display.set_mode(size)

# 제목 설정
title = 'PAC MAN'
pygame.display.set_caption(title)

# 게임 내 설정
clock = pygame.time.Clock()
color = (120,120,255)

# 객체 생성
class obj:
    # 초기 위치 설정
    def __init__(self):
        self.x = 0
        self.y = 0
    # 이미지 추가
    def put_img(self, address):
        self.img = pygame.image.load(address).convert_alpha()
        self.sx, self.sy = self.img.get_size()
    # 이미지 사이즈 변경
    def change_size(self, sx, sy):
        self.img = pygame.transform.scale(self.img,(sx,sy))
        self.sx, self.sy = self.img.get_size()
    # 객체 그리기
    def show(self):
        screen.blit(self.img, (self.x, self.y))
 
# 팩맨 객체 생성       
pacman = obj()
pacman.put_img("./이미지/팩맨1 위쪽.png")
pacman.change_size(40,40)
pacman.x = (size[0]//2) - (pacman.sx//2)
pacman.y = (size[1]*2)//3
        
# 게임 루프
running = True
while running:

    # FPS 설정
    clock.tick(60)

    # event가 일어났는지 확인
    for event in pygame.event.get():
        # 창 X표시 누르면 게임 종료
        if event.type == pygame.QUIT:
            running = False
    # 화면 채우기
    screen.fill(color)
    
    # 팩맨 그리기
    pacman.show()

    # 화면을 매 프레임마다 업데이트
    pygame.display.flip()

pygame.quit()
```

# 팩맨 이동

```python
import pygame

# Pygame 초기화
pygame.init()

# 게임 화면 설정
screen_width = 1000
screen_height = 1000
size = [screen_width,screen_height]
screen = pygame.display.set_mode(size)

# 제목 설정
title = 'PAC MAN'
pygame.display.set_caption(title)

# 게임 내 설정
clock = pygame.time.Clock()
color = (120,120,255)

# 객체 생성
class obj:
    # 초기 위치 설정
    def __init__(self):
        self.x = 0
        self.y = 0
    # 이미지 추가
    def put_img(self, address):
        self.img = pygame.image.load(address).convert_alpha()
        self.sx, self.sy = self.img.get_size()
    # 이미지 사이즈 변경
    def change_size(self, sx, sy):
        self.img = pygame.transform.scale(self.img,(sx,sy))
        self.sx, self.sy = self.img.get_size()
    # 객체 그리기
    def show(self):
        screen.blit(self.img, (self.x, self.y))
 
# 팩맨 객체 생성       
pacman = obj()
pacman.put_img("./이미지/팩맨1 위쪽.png")
pacman.change_size(40,40)
pacman.x = (size[0]//2) - (pacman.sx//2)
pacman.y = (size[1]*2)//3
pacman.move = 8

# 팩맨 객체 이동방향 초기 값
left_go = False
right_go = False
up_go = False
down_go = False

# 필요한 변수 선언
key = [pygame.K_LEFT,pygame.K_RIGHT,pygame.K_UP,pygame.K_DOWN]
image = ["./이미지/팩맨1 왼쪽.png","./이미지/팩맨1 오른쪽.png","./이미지/팩맨1 위쪽.png","./이미지/팩맨1 아래쪽.png"]

# 게임 루프
running = True
while running:

    # FPS 설정
    clock.tick(60)

    # event가 일어났는지 확인
    for event in pygame.event.get():
        # 창 X표시 누르면 게임 종료
        if event.type == pygame.QUIT:
            running = False
        # 키 입력 확인
        if event.type == pygame.KEYDOWN:
            for i in range(4):
                if event.key == key[i]:
                    go = [1 if i == j else 0 for j in range(4)]
                    left_go = bool(go[0])
                    right_go = bool(go[1])
                    up_go = bool(go[2])
                    down_go = bool(go[3])
                    pacman.put_img(image[i]) # 이미지 재설정
                    pacman.change_size(40,40)
    
    # 각 이동 방향으로 이동
    if left_go == True:
        pacman.x -= pacman.move
        if pacman.x <= 0:
            pacman.x = 0
    if right_go == True:
        pacman.x += pacman.move
        if pacman.x >= size[0] - pacman.sx:
            pacman.x = size[0] - pacman.sx
    if up_go == True:
        pacman.y -= pacman.move
        if pacman.y <= 0:
            pacman.y = 0
    if down_go == True:
        pacman.y += pacman.move
        if pacman.y >= size[1] - pacman.sy:
            pacman.y = size[1] - pacman.sy

    # 화면 채우기
    screen.fill(color)
    
    # 팩맨 그리기
    pacman.show()

    # 화면을 매 프레임마다 업데이트
    pygame.display.flip()

pygame.quit()
```
