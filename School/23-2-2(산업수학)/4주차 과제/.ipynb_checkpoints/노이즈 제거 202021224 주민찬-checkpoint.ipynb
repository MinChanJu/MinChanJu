# 1. 사인 파장 2개를 합쳐 파장 하나로 합침

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
time = np.arange(0,1,0.001)

sin1 = 5*np.sin(2*np.pi*4*time)
sin2 = 2*np.sin(2*np.pi*5*time) + 1

wave = sin1 + sin2
```

```python
plt.subplot(3,1,1)
plt.plot(time, sin1)
plt.title('sin1')

plt.subplot(3,1,2)
plt.plot(time, sin2)
plt.title('sin2')

plt.subplot(3,1,3)
plt.plot(time, wave)
plt.title('wave')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import soundfile as sf

# 오디오 파일로 저장
output_file_path = "wave_audio.wav"
sf.write(output_file_path, wave, 100)
```

# 2. 노이즈를 파장에 추가

```python
noise = np.random.uniform(0, 5, size=len(time))
n_wave = wave + noise
```

```python
plt.subplot(3,1,1)
plt.plot(time, n_wave)
plt.title('n_wave')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import soundfile as sf

# 오디오 파일로 저장
output_file_path = "n_wave_audio.wav"
sf.write(output_file_path, n_wave, 100)
```

# 3. 노이즈 제거
### 3-1 푸리에 변환
### 3-2 푸리에 변환한 값이 150 이하인 값들을 0으로 바꿈(노이즈 제거)
### 3-3 역푸리에 변환

```python
fft_wave = abs(np.fft.fft(n_wave))
re_wave = np.where(fft_wave <= 150, 0, fft_wave)
ifft_wave = np.fft.ifft(re_wave)
```

```python
plt.subplot(2,1,1)
plt.plot(time, fft_wave)
plt.title('fft_wave')

plt.subplot(2,1,2)
plt.plot(time, re_wave)
plt.title('re_wave')

plt.tight_layout()
plt.show()
```

```python
plt.subplot(3,1,1)
plt.plot(time, ifft_wave)
plt.title('ifft_wave')

plt.subplot(3,1,2)
plt.plot(time, wave)
plt.title('wave')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import soundfile as sf

# 오디오 파일로 저장
output_file_path = "ifft_wave_audio.wav"
sf.write(output_file_path, abs(np.fft.ifft(re_wave)), 100)
```

# 4. 실습
### 4-1 오디오 데이터 생성
### 4-2 노이즈를 임의로 생성후 결합
### 4-3 푸리에 변환
### 4-4 입계값 이하의 주파수 제거
### 4-5 역푸리에 변환

```python
import numpy as np
from scipy.fft import fft, ifft
from scipy.io import wavfile

# 오디오 생성
duration_in_seconds = 5  # 생성할 오디오의 길이 (초)
sample_rate = 44100  # 샘플링 속도
frequency = 440  # 주파수 (Hz)

# 오디오 데이터 생성 (사인 파형)
t = np.linspace(0, duration_in_seconds, int(sample_rate * duration_in_seconds), endpoint=False)
audio_data = np.sin(2 * np.pi * frequency * t)

# 노이즈 생성 (임의로 생성)
noise = 0.2* np.random.normal(size=len(t))

# 오디오 데이터와 노이즈 결합
audio_data_with_noise = audio_data + noise

# 원본 오디오 데이터를 파일로 저장 (예: original_audio.wav)
original_output_file_path = "original_audio.wav"
wavfile.write(original_output_file_path, sample_rate, np.int16(audio_data_with_noise * 32767))

# 푸리에 변환을 수행하여 주파수 도메인으로 이동
fft_data = fft(audio_data_with_noise)

# 노이즈를 식별하여 임계값 이하의 주파수를 제거
threshold = 100
fft_data[np.abs(fft_data) < threshold] = 0

# 역 푸리에 변환을 수행하여 시간 도메인으로 이동
filtered_audio_data = ifft(fft_data)

# 결과를 정수로 변환
filtered_audio_data = np.int16(filtered_audio_data.real)

# 오디오 데이터를 파일로 저장 (예: filtered_output.wav)
filtered_output_file_path = "filtered_output.wav"
wavfile.write(filtered_output_file_path, sample_rate, filtered_audio_data)
```
