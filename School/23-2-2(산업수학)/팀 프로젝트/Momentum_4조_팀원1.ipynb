## 최적화
- 정의 : 손실함수의 값을 최소화하는 파라미터를 구하는 과정
- 손실함수
    - 딥러닝에서 인공지능이 예측하는 값과 실제 값의 차이
    - 즉, 인공지능을 피드백하기 위한 함수
- 최적화는 이 손실함수를 최소화 시키는 것을 목적으로 한다

## Gradient descent
- 수치 최적화는 경사하강법을 기본 개념으로 이용
- 경사하강법은 함수의 기울기 벡터에 의존하여 최솟값을 찾아내는 방법
- $F(x)_{n+1}=f(x)_n-a*grad(f(x)_n)$
- a는 학습률(learning rate)

### 한계점
1. 학습률에 민감하다
2. 국소 최솟값(local minimum)에 빠지기 쉽다
3. 안장점(saddle point)에 벗어날 수 없다

#### 학습률에 민감하다

```python
import numpy as np
import matplotlib.pyplot as plt

# 함수 정의
def func(x):
    return x**4 - 4*x

def grad_func(x):
    return 4*x**3 - 4

# x 값 생성
x_values = np.linspace(-1, 3, 400)
y_values = func(x_values)

# 그래프 그리기
plt.plot(x_values, y_values, label='y=x^4-4x')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graph of y=x^4-4x')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def func(x):
    return x**4 - 4 * x

def grad_func(x):
    return 4 * x**3 - 4

def gradient_descent(starting_point, learning_rate, iterations):
    x_values = []
    x = starting_point
    
    for i in range(iterations):
        gradient_value = grad_func(x)
        x = x - learning_rate * gradient_value
        x_values.append(x)
        
        # 출력 추가
        print(f"Iteration {i+1}: x_{i+1} = {x:.5f}")
        
    return x_values

# 초기값, 학습률, 반복 횟수 설정
initial_point = 2
learning_rate = 0.01
iterations = 10

# gradient descent 실행
x_values = gradient_descent(initial_point, learning_rate, iterations)

# 결과 출력
min_point = x_values[-1]
print(f"\n최솟값을 찾은 x 값: {min_point:.5f}")

# 시각화
x_range = np.linspace(-1.5, 3.5, 100)
plt.plot(x_range, func(x_range), label='f(x) = x^4 - 4x')
plt.scatter(x_values, [func(x) for x in x_values], color='red', label='Gradient Descent Path')
plt.title('Gradient Descent Optimization')
plt.xlabel('x')
plt.ylabel('f(x)')

# 각 점에 이름 추가
for i, x in enumerate(x_values):
    plt.text(x, func(x), f'$x_{i+1}$', fontsize=8, ha='right', va='bottom')

plt.legend()
plt.grid(True)
plt.show()
```

#### Saddle point를 벗어나지 못한다

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 함수 정의
def func(x, y):
    return x**2 - y**2

# 편미분 함수 정의
def partial_derivative_x(x, y):
    return 2 * x

def partial_derivative_y(x, y):
    return -2 * y

# 경사하강법 함수 정의
def gradient_descent(x, y, learning_rate, num_iterations):
    path = [(x, y, func(x, y))]  # 경로 저장을 위한 리스트

    for _ in range(num_iterations):
        x = x - learning_rate * partial_derivative_x(x, y)
        y = y - learning_rate * partial_derivative_y(x, y)
        path.append((x, y, func(x, y)))

    return path

# 초기값 설정
initial_x = 80.0
initial_y = -1.0
learning_rate = 0.1
num_iterations = 20

# 경사하강법 경로 계산
path = gradient_descent(initial_x, initial_y, learning_rate, num_iterations)
path = np.array(path)

# 2D 그리드 생성
x = np.linspace(-100, 100, 1000)
y = np.linspace(-50, 50, 1000)
X, Y = np.meshgrid(x, y)
Z = func(X, Y)

# 3D 그래프 생성
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# 표면 플로팅
ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.5, rstride=10, cstride=10)

# 경로 플로팅
ax.plot(path[:, 0], path[:, 1], path[:, 2], color='red', marker='o')

# 마지막 점의 좌표 출력
final_x, final_y, final_z = path[-1]
print(f"Final Point: ({final_x}, {final_y}, {final_z})")

# 축 레이블 추가
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('3D Plot with Gradient Descent Path')

# 그래프 표시
plt.show()
```

- 안장점은 한 방향에서는 극소이지만 한 방향에서는 극대인 지점을 의미한다
- 안장점에서의 gradient 값이 0이므로 점이 더이상 업데이트 되지 않고 안장점에 국한된다
- 위 함수 $z=x^2-y^2$은 (0,0,0)에서 안장점을 가진다
- 편미분 결과 x축에서는 극소점이고 y축에서는 극대점이므로 경사하강법을 하면 x좌표는 0에 수렴하지만 y좌표는 절댓값이 커진다
- 하지만 초기 y좌표가 0이라면 편미분값이 0이므로 y좌표가 안장점에서 벗어나지 못한다

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 함수 정의
def func(x, y):
    return x**2 - y**2

# 편미분 함수 정의
def partial_derivative_x(x, y):
    return 2 * x

def partial_derivative_y(x, y):
    return -2 * y

# 경사하강법 함수 정의
def gradient_descent(x, y, learning_rate, num_iterations):
    path = [(x, y, func(x, y))]  # 경로 저장을 위한 리스트

    for _ in range(num_iterations):
        x = x - learning_rate * partial_derivative_x(x, y)
        y = y - learning_rate * partial_derivative_y(x, y)
        path.append((x, y, func(x, y)))

    return path

# 초기값 설정
initial_x = 80.0
initial_y = 0
learning_rate = 0.1
num_iterations = 20

# 경사하강법 경로 계산
path = gradient_descent(initial_x, initial_y, learning_rate, num_iterations)
path = np.array(path)

# 2D 그리드 생성
x = np.linspace(-100, 100, 1000)
y = np.linspace(-50, 50, 1000)
X, Y = np.meshgrid(x, y)
Z = func(X, Y)

# 3D 그래프 생성
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# 표면 플로팅
ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.5, rstride=10, cstride=10)

# 경로 플로팅
ax.plot(path[:, 0], path[:, 1], path[:, 2], color='red', marker='o')

# 마지막 점의 좌표 출력
final_x, final_y, final_z = path[-1]
print(f"Final Point: ({final_x}, {final_y}, {final_z})")

# 축 레이블 추가
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('3D Plot with Gradient Descent Path')

# 그래프 표시
plt.show()
```

## Momentum

- 함수는 convex와 non-convex 함수로 나뉜다
- convex함수는 경사하강법을 통해 최적의 값으로 갈 수 있다
- non-convex함수는 local minimum이 존재하기 때문에 최적의 값으로 가지 못하고 local minimum에 빠지기 쉽다

- momentum은 경사하강법을 통해 최소화시키는 것에 관성을 부여하는 방법
- 이전 이동 방향을 기억하여 기울기의 크기를 비교하고 어느정도 추가로 이동

$v_t=a*v_{t-1}+b*grad(f(v_{t-1}))$

a=momentum coefficient b=learning rate 일반적으로 관성계수는 0.9로 설정

```python
import numpy as np
import matplotlib.pyplot as plt

def func2(x):
    return x**4 + 8*x**3 - 6*x**2 + 2*x

def grad_func2(x):
    return 4*x**3 + 24*x**2 - 12*x + 2

def momentum_gradient_descent(starting_point, learning_rate, momentum, iterations):
    x_values = []
    x = starting_point
    v = 0  # Momentum term

    for i in range(iterations):
        gradient_value = grad_func2(x)
        v = momentum * v - learning_rate * gradient_value
        x = x + v
        x_values.append(x)

        # 출력 추가
        print(f"Iteration {i + 1}: x_{i + 1} = {x:.5f}")

    return x_values

# 초기값, 학습률, 모멘텀, 반복 횟수 설정
initial_point = 1
learning_rate = 0.01
momentum = 0.9
iterations = 10

# momentum 기법으로 gradient descent 실행
x_values_momentum = momentum_gradient_descent(initial_point, learning_rate, momentum, iterations)

# 결과 출력
min_point_momentum = x_values_momentum[-1]
print(f"\n최솟값을 찾은 x 값 (Momentum): {min_point_momentum:.5f}")

# 시각화
x_range = np.linspace(-7, 2, 1000)
plt.plot(x_range, func2(x_range), label='f(x) = x^4 + 8x^3 - 6x^2 + 2x')
plt.scatter(x_values_momentum, [func2(x) for x in x_values_momentum], color='blue', label='Momentum Gradient Descent Path')
plt.title('Momentum Optimization with Local Minimum')
plt.xlabel('x')
plt.ylabel('f(x)')

# 각 점에 이름 추가
for i, x in enumerate(x_values_momentum):
    plt.text(x, func2(x), f'$x_{i + 1}={x:.5f}$', fontsize=8, ha='right')

plt.legend()
plt.grid(True)
plt.show()
```
