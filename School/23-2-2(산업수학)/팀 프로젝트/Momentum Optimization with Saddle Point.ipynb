Momentum Optimization with Saddle Point

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def momentum_optimizer(theta, gradient, learning_rate, momentum, prev_velocity):
    velocity = momentum * prev_velocity + (1 - momentum) * gradient
    theta = theta - learning_rate * velocity
    return theta, velocity

# 목적 함수 정의 (3차원 saddle point를 가정)
def objective_function(x):
    return x[0]**2 - x[1]**2

# 초기 파라미터와 하이퍼파라미터 설정
theta_initial = [-4,0]  # 임의의 초기값
learning_rate = 0.5
momentum = 0.9
prev_velocity = np.zeros_like(theta_initial)  # 초기 모멘텀 벡터

num_steps = 1000 # 최적화를 100번 반복

# 경사 하강법을 여러 단계로 반복
theta_history = [theta_initial]
for step in range(num_steps):
    # 목적 함수의 기울기 계산
    gradient = np.array([2*theta_initial[0] - theta_initial[1], 2*theta_initial[1] - theta_initial[0]])

    # 모멘텀 최적화 알고리즘 적용
    theta_initial, prev_velocity = momentum_optimizer(theta_initial, gradient, learning_rate, momentum, prev_velocity)

    # 업데이트된 파라미터 기록
    theta_history.append(theta_initial)

# 결과 시각화
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# 목적 함수의 등고선 플로팅
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)
Z = objective_function([X, Y])
ax.contour(X, Y, Z, levels=50, cmap='viridis', alpha=0.5)

# 최적화 경로 플로팅
theta_history = np.array(theta_history)
ax.plot(theta_history[:, 0], theta_history[:, 1], objective_function(theta_history.T), 'ro-', label='Optimization Path')

print(theta_initial)

ax.set_title('Momentum Optimization with Saddle Point')
```

```python
def objective_function(x, y):
    return x**2 + y**2
```
